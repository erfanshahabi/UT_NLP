# پاسخ خود را اینجا بنویسید:
1. pos_tags (Part of Speech Tags):
این ستون شامل برچسب‌های نحوی هر کلمه در جمله است و مشخص می‌کند که هر کلمه چه نقشی در جمله دارد (مانند اسم، فعل، قید و غیره).

مثال مقادیر:

NOUN: اسم
VERB: فعل
ADJ: صفت
ADV: قید
کاربرد:

این برچسب‌ها برای تحلیل نحوی جمله و بررسی ساختار گرامری استفاده می‌شوند.
به‌عنوان یک ویژگی مفید در وظایفی مثل تحلیل احساسات یا شناسایی موجودیت‌ها.

2. chunk_tags (Chunking Tags):
این ستون مشخص می‌کند که کلمات در چه گروه‌های نحوی (مانند گروه اسمی یا گروه فعلی) قرار می‌گیرند.

مثال مقادیر:

NP: گروه اسمی (Noun Phrase)
VP: گروه فعلی (Verb Phrase)
PP: گروه حرف اضافه‌ای (Prepositional Phrase)
کاربرد:
بریا درک ساختار جمله به صورت گروهی (به جای کلمات منفرد) مثرلا برای شناسایی عبارات اسمی که ممکن است بک موجودیت خاص باشند.

3. ner_tags (Named Entity Recognition Tags): این ستون شامل برچسب‌های موجودیت‌های نام‌برده است که برای شناسایی و دسته‌بندی موجودیت‌ها به‌کار می‌رود. این موجودیت‌ها ممکن است اشخاص، مکان‌ها، سازمان‌ها و غیره باشند.

مثال مقادیر:

B-PER: ابتدای موجودیت شخصی (Begin Person)
I-PER: داخل موجودیت شخصی (Inside Person)
B-ORG: ابتدای موجودیت سازمانی (Begin Organization)
I-ORG: داخل موجودیت سازمانی (Inside Organization)
O: بدون موجودیت (Outside)
کاربرد: این برچسب‌ها برای شناسایی موجودیت‌ها در متن و تجزیه‌وتحلیل اطلاعات خاص (مانند نام‌ها، مکان‌ها و سازمان‌ها) استفاده می‌شوند. این ویژگی برای وظایف استخراج اطلاعات و پردازش متن مفید است.

4. id: این ستون شامل شناسه‌های منحصر به فرد برای هر نمونه است. شناسه‌ها به تفکیک نمونه‌ها و ردیابی داده‌ها کمک می‌کنند.

مثال مقادیر:

یک رشته از کاراکترها که هر نمونه را شناسایی می‌کند.
کاربرد: این شناسه‌ها برای مدیریت و پیگیری نمونه‌ها در طول پردازش داده‌ها یا آموزش مدل‌ها استفاده می‌شوند.

5. tokens: این ستون شامل لیستی از کلمات (توکن‌ها) در هر جمله است.

مثال مقادیر:

["John", "lives", "in", "New", "York"]
کاربرد: این ویژگی برای پردازش زبانی و انجام وظایفی مانند توکن‌سازی، تحلیل نحوی و شناسایی موجودیت‌ها استفاده می‌شود.

این کد یک کلاس به نام Vocab ایجاد می‌کند که برای ساخت واژه‌نامه (Vocabulary) از یک دیتاست استفاده می‌کند. این کلاس توکن‌ها را از داده‌های ورودی استخراج کرده، آنها را فیلتر می‌کند و سپس واژه‌نامه‌ای با استفاده از توکن‌ها و شناسه‌های مربوطه می‌سازد. همچنین، این کلاس توابعی برای تبدیل توکن‌ها به شناسه و بالعکس فراهم می‌کند.
ورودی این کلاس دیتا ست است و مقدار کاتاف را برابر با متن سوال برابر ۵ گرفتم.
سپس تابع build vocab را تعریف کردم که این تابع از دیتا ست توکن ها را استخراج کرده و شمارش میکند و توکن هایی که تعداد آن ها در دیتا ست بیشتر از ۵ یا همان کات اف هست را نگه میدارد.
در مرحله بعد تایع token_to_id تعریف شده که این تابع یک توکن را گرفتع و شناسه مربوط به آن را باز میگرداند و اگر توکن در وکب وجود نداشته باشد شناسه <UNK> بازگردادنده می شود.
تابع بعدی تایع id_to_token است که این تابع شناسه را گرافته و توکن مربوط به آن شناسه را باز میگرداند. vocab_size نیز اندازه واژه نامه را باز میگرداند.
بعد از تعریف واژه نامه ابتدا دیتا ست را بارگزاری می کنیم و سپس یک شی از کلاس وکب با پارامتر کات اف برابر ۵ ساخته می شود و توکن های ویژه را به وکب اضافه میکنیم و یک نمونه از تبیل توکن به شناسه و شناسه به توکن را نمایش می دهیم.
به عنوان خروجی و برای اینکه از روند صحیح برنامه مطلع شویم انداره واژه نامه را چاپ میکنیم و شناسه مربوط به توکن <s> و بالعکس را چاپ میکنیم.


```python
from collections import Counter
from datasets import load_dataset

class Vocab:
    def __init__(self, dataset, cutoff=5):
        self.cutoff = cutoff
        self.token_to_id_dict = {}
        self.id_to_token_dict = {}
        self.build_vocab(dataset)
        
    def build_vocab(self, dataset):
    
        all_tokens = []
        for example in dataset["train"]:
            all_tokens.extend(example["tokens"])  
        
        token_counts = Counter(all_tokens)
        
        filtered_tokens = [token for token, count in token_counts.items() if count >= self.cutoff]
        
        special_tokens = ['<pad>', '<unk>', '<s>', '</s>']
        filtered_tokens = special_tokens + filtered_tokens
        
        self.token_to_id_dict = {token: idx for idx, token in enumerate(filtered_tokens)}
        self.id_to_token_dict = {idx: token for token, idx in self.token_to_id_dict.items()}
    
    def token_to_id(self, token):
        return self.token_to_id_dict.get(token, self.token_to_id_dict['<unk>'])  
    
    def id_to_token(self, idx):
        return self.id_to_token_dict.get(idx, '<unk>') 
    def vocab_size(self):
        return len(self.token_to_id_dict)


dataset = load_dataset("eriktks/conll2003")

vocab = Vocab(dataset, cutoff=10) 


print(f"Vocabulary size: {vocab.vocab_size()}")


token_example = "<s>"
token_id = vocab.token_to_id(token_example)
print(f"Token: {token_example} -> ID: {token_id}")
print(f"ID: {token_id} -> Token: {vocab.id_to_token(token_id)}")
```

هدف از این کد تبدیل داده های متنی به فرمتی قابل پردازش برای مدل ماست که در این فرایند داده ها به صورت batch سازماندهی میشوند و توکن ها و برچسب های NER در هر جمله به طول های یکسان با استفاده از padding تنظیم می شوند.
به همین منظور تابع collate_fn را تعریف کردم که برای تبدیل و آماده سازی هر دسته طراحی شده است.
ووردی این تابع batch است که یک لیست شامل تعدادی از داده ها به اندازه batch سایز می باشد.
سپس هر کلمه در جمله از طریق وکب به شناسه عددی متناظرش تنبدیل می شود.
در مرحله بعد برچسب های NER را برای هر توکن در جمله به شناسه عددی متناظر تبدیل می کنیم.
سپس جملات با طول های مختلف را با مقدار <pad> تا طول یکسان پر می کنیم.
برچسب های جملات نیز با مفداز -۱ پر می شوند تا طول ان ها یکسان شود.
خرجی این تابع padded_tokens , padded_lables است که ماتریسی هستند شامب شناسه های توکن های جملات به صورت دسته ای و برچسب های NER.
در مرحله بعدی از dataloader برای بارگذاری دسته ای داده ها استفاده شده است. پارامترهای این بخش dataset['train'] , dataset['validation'] هستند که داده های اموزشی و تست می باشند.
batch_Size طبق گفته سوال برابر ۳۲ تنظیم شده است.

```python
from torch.nn.utils.rnn import pad_sequence
import torch

def collate_fn(batch):
    tokens = [torch.tensor([vocab.token_to_id(token) for token in example['tokens']], dtype=torch.long) for example in batch]

    labels = [torch.tensor(example['ner_tags'], dtype=torch.long) for example in batch]
    padded_tokens = pad_sequence(tokens, batch_first=True, padding_value=vocab.token_to_id('<pad>'))
    padded_labels = pad_sequence(labels, batch_first=True, padding_value=-1) 

    return padded_tokens, padded_labels
from torch.utils.data import DataLoader

train_loader = DataLoader(dataset['train'], batch_size=32, collate_fn=collate_fn)
val_loader = DataLoader(dataset['validation'], batch_size=32, collate_fn=collate_fn)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report
import numpy as np

class Encoder(nn.Module):
    def __init__(self, vocab_size, embedding_size, hidden_size):
        super(Encoder, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_size)
        self.rnn = nn.RNN(embedding_size, hidden_size, bidirectional=True, batch_first=True)
        self.fc = nn.Linear(2 * hidden_size, vocab_size)

    def forward(self, x):
        embedded = self.embedding(x)
        rnn_out, _ = self.rnn(embedded)
        out = self.fc(rnn_out)
        return out


vocab_size = vocab.vocab_size()  
embedding_size = 64
hidden_size = 64
batch_size = 32
num_epochs = 10
learning_rate = 1e-3


model = Encoder(vocab_size, embedding_size, hidden_size)

optimizer = optim.Adam(model.parameters(), lr=learning_rate)
criterion = nn.CrossEntropyLoss(ignore_index=-1)  
train_loss_values = []
train_accuracy_values = []
val_loss_values = []
val_accuracy_values = []


for epoch in range(num_epochs):
    model.train()
    epoch_train_loss = 0
    epoch_train_correct = 0
    epoch_train_total = 0
    for batch in train_loader:
        tokens, labels = batch
        optimizer.zero_grad()
        outputs = model(tokens)
        loss = criterion(outputs.view(-1, vocab_size), labels.view(-1))
        loss.backward()
        optimizer.step()
        
        epoch_train_loss += loss.item()
        predicted_labels = torch.argmax(outputs, dim=-1)
        epoch_train_correct += (predicted_labels == labels).sum().item()
        epoch_train_total += labels.numel()

    train_loss_values.append(epoch_train_loss / len(train_loader))
    train_accuracy_values.append(epoch_train_correct / epoch_train_total)
    print(f"Epoch {epoch+1}, Train Loss: {epoch_train_loss / len(train_loader)}, Train Accuracy: {epoch_train_correct / epoch_train_total}")
    
    model.eval()
    epoch_val_loss = 0
    epoch_val_correct = 0
    epoch_val_total = 0
    predictions = []
    true_labels = []
    
    with torch.no_grad():
        for batch in val_loader:
            tokens, labels = batch
            outputs = model(tokens)
            loss = criterion(outputs.view(-1, vocab_size), labels.view(-1))
            epoch_val_loss += loss.item()
          
            predicted_labels = torch.argmax(outputs, dim=-1)
            epoch_val_correct += (predicted_labels == labels).sum().item()
            epoch_val_total += labels.numel()
          
            predictions.extend(predicted_labels.view(-1).cpu().numpy())
            true_labels.extend(labels.view(-1).cpu().numpy())

    valid_indices = np.array(true_labels) != -1  
    filtered_true_labels = np.array(true_labels)[valid_indices]
    filtered_predictions = np.array(predictions)[valid_indices]
 
    val_loss_values.append(epoch_val_loss / len(val_loader))
    val_accuracy_values.append(epoch_val_correct / epoch_val_total)
    print(f"Epoch {epoch+1}, Validation Loss: {epoch_val_loss / len(val_loader)}, Validation Accuracy: {epoch_val_correct / epoch_val_total}")

print("\nFinal Classification Report on Validation Data:")
print(classification_report(filtered_true_labels, filtered_predictions, zero_division=1))


plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(train_loss_values, label='Train Loss')
plt.plot(val_loss_values, label='Validation Loss')
plt.title('Loss Curve')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(train_accuracy_values, label='Train Accuracy')
plt.plot(val_accuracy_values, label='Validation Accuracy')
plt.title('Accuracy Curve')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
from datasets import load_dataset


dataset = load_dataset("EdinburghNLP/xsum")

from sklearn.model_selection import train_test_split

train_data = dataset['train']
validation_data = dataset['validation']

train_sample = train_data.select(range(int(len(train_data) * 0.3)))

validation_sample = validation_data.select(range(int(len(validation_data) * 0.1)))

train_input = [item['document'] for item in train_sample]
train_target = [item['summary'] for item in train_sample]

val_input = [item['document'] for item in validation_sample]
val_target = [item['summary'] for item in validation_sample]
```

```python
import re


def tokenize(text):
  
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)
    tokens = text.split()
    
    return tokens

train_input_tokens = [tokenize(doc) for doc in train_input]
train_target_tokens = [tokenize(summary) for summary in train_target]

val_input_tokens = [tokenize(doc) for doc in val_input]
val_target_tokens = [tokenize(summary) for summary in val_target]
```

برای این که مطمئن شوم که کد عملکرد مناسب را داشته است یا نه اندازه وکب و ۱۰ توکن اول و شناسه مربوط به آن ها را چاپ کردم

```python
from collections import Counter

class Vocab:
    def __init__(self, dataset, cutoff=10):
        self.cutoff = cutoff
        self.token_to_id_dict = {}
        self.id_to_token_dict = {}
        self.build_vocab(dataset)
        
    def build_vocab(self, dataset):
     
        all_tokens = []
        
        for tokens in dataset:
            all_tokens.extend(tokens)  
        
  
        token_counts = Counter(all_tokens)
        
        filtered_tokens = [token for token, count in token_counts.items() if count >= self.cutoff]
        
        special_tokens = ['<pad>', '<unk>', '<s>', '</s>']
        filtered_tokens = special_tokens + filtered_tokens
        
        self.token_to_id_dict = {token: idx for idx, token in enumerate(filtered_tokens)}
        self.id_to_token_dict = {idx: token for token, idx in self.token_to_id_dict.items()}
    
    def token_to_id(self, token):
        return self.token_to_id_dict.get(token, self.token_to_id_dict['<unk>'])  
    
    def id_to_token(self, idx):
        return self.id_to_token_dict.get(idx, '<unk>')  
    def vocab_size(self):
        return len(self.token_to_id_dict)


vocab = Vocab(train_input_tokens + train_target_tokens)

print(f"Vocabulary size: {vocab.vocab_size()}")

for token, idx in list(vocab.token_to_id_dict.items())[:10]:
    print(f"Token: {token} -> ID: {idx}")
```

```python
import numpy as np

glove_file_path = "/kaggle/input/glove-100-2/glove.6B.100d.txt"


def load_glove_embeddings(file_path):
    embeddings_index = {}
    with open(file_path, "r", encoding="utf-8") as f:
        for line in f:
            values = line.split()
            word = values[0]
            vector = np.array(values[1:], dtype="float32")
            embeddings_index[word] = vector
    return embeddings_index

glove_embeddings = load_glove_embeddings(glove_file_path)

print(f"Loaded {len(glove_embeddings)} word vectors.")
```

در این قسمت از کد دیتاست و دیتا لودری برای تبدیل داده ها به فرمت های مناسب برای ورودی و خروجی مدل ساخته ایم.
ابتدا کلاس seqtoseqdataset تعرف شده است  که از کلاس torch.utils.data.Dataset ارث بری می‌کند. این کلاس برای مدیریت داده‌های ورودی و خروجی استفاده می‌شود. سپس تابع encode تعریف شده است که در این تابع، هر جمله به یک دنباله از شناسه‌های عددی تبدیل می‌شود. برای این کار، از دیکشنری vocab که در آن کلمات به شناسه‌های عددی نگاشت شده‌اند، استفاده می‌شود.
متد len تعداد کل داده‌های موجود در مجموعه را بر می‌گرداند و متد getitem را تعریف کردم. در این متد، داده‌های ورودی و هدف برای یک نمونه خاص (با استفاده از ایندکس idx) بر اساس دیکشنری vocab به شناسه‌های عددی تبدیل می‌شوند.
سپس طول هر دنباله (input_seq و target_seq) بر اساس max_len تنظیم می‌شود. اگر طول دنباله کمتر از max_len باشد، با <pad> پر می‌شود. سپس تابع collate_fn را تعریف کرده ام. این تابع برای تنظیم نحوه‌ی گروه‌بندی نمونه‌ها در هر بچ (batch) استفاده می‌شود.
torch.cat به این منظور استفاده می‌شود که به هر دنباله input_seq و target_seq علامت‌های شروع (<s>) و پایان (</s>) اضافه کند.
سپس، pad_sequence از PyTorch برای پر کردن دنباله‌ها به طول ثابت (بیشترین طول دنباله در بچ) با استفاده از مقدار پرکننده (<pad>) استفاده می‌شود. در مرحله بعد یک شیء DataLoader از PyTorch ساخته می‌شود که برای بارگذاری داده‌ها در بچ‌های اندازه 32 به صورت تصادفی استفاده خواهد شد.
تابع collate_fn برای فرآیند گروه‌بندی داده‌ها به کار می‌رود.
و در نهایت برای اطمینان از عملکرد کد ابعاد ورودی و هدف در یک بچ را چاپ کردم.

```python
import torch
from torch.utils.data import DataLoader
from torch.nn.utils.rnn import pad_sequence
import torch.nn.functional as F

class Seq2SeqDataset(torch.utils.data.Dataset):
    def __init__(self, input_data, target_data, vocab, max_len=50):
        self.input_data = input_data
        self.target_data = target_data
        self.vocab = vocab
        self.max_len = max_len
        
    def encode(self, sentence):
        return [self.vocab.token_to_id(token) for token in sentence]

    def __len__(self):
        return len(self.input_data)
    
    def __getitem__(self, idx):
        input_seq = self.encode(self.input_data[idx])
        target_seq = self.encode(self.target_data[idx])
        
        input_seq = input_seq[:self.max_len] + [self.vocab.token_to_id('<pad>')] * (self.max_len - len(input_seq))
        target_seq = target_seq[:self.max_len] + [self.vocab.token_to_id('<pad>')] * (self.max_len - len(target_seq))
        
        return torch.tensor(input_seq), torch.tensor(target_seq)

def collate_fn(batch):
    input_seqs, target_seqs = zip(*batch)
    input_seqs = [torch.cat([torch.tensor([vocab.token_to_id('<s>')]), seq, torch.tensor([vocab.token_to_id('</s>')])]) for seq in input_seqs]
    target_seqs = [torch.cat([torch.tensor([vocab.token_to_id('<s>')]), seq, torch.tensor([vocab.token_to_id('</s>')])]) for seq in target_seqs]

    input_seqs = pad_sequence(input_seqs, batch_first=True, padding_value=vocab.token_to_id('<pad>'))
    target_seqs = pad_sequence(target_seqs, batch_first=True, padding_value=vocab.token_to_id('<pad>'))
    return input_seqs, target_seqs



train_data_loader = DataLoader(Seq2SeqDataset(train_input_tokens, train_target_tokens, vocab),
                               batch_size=32,  
                               collate_fn=collate_fn)

for input_seqs, target_seqs in train_data_loader:
    print(f"Batch input shape: {input_seqs.shape}")
    print(f"Batch target shape: {target_seqs.shape}")
    break 
```

هدف از این کد ساخت یک شبکه encoder-decoder با توجه به مکانیزم attention است. این شبکه شامل سه لایه امبدنیگ، bidirectional GRU و لایه خطی است.
در این کد پس از ایمپور کردن کتابخانه های مورد نیاز کلاس attention را تعریف کرده ایم. این کلاس یک مکانیزم ساده attention را پیاده سازی می کند که اجازه می دهد مدل به قسمت های خاصی از ورودی که اینجا encoder است دسترسی پیدا کند.
در این کلاس درون تابع __init__ یک لایه Linear برای تولید وزن‌های توجه (attention weights) و یک لایه Softmax برای نرمال‌سازی این وزن‌ها به کار می‌رود. و در تابع forward، وزن‌های توجه محاسبه شده و سپس از آنها برای محاسبه context استفاده می‌شود.
سپس کلاس seq2seq تعریف شده است که یک شبکه encoder-decoder می باشد.
این کلاس شامل سه لایه اصلی زیر می باشد:
Layer: ابتدا ورودی‌ها به فضای برداری تبدیل می‌شوند. برای این کار از nn.Embedding استفاده می‌شود که یک embedding matrix را به عنوان ورودی می‌گیرد.
GRU Layer: سپس ورودی‌های تبدیل شده به امبدینگ های برداری از طریق یک لایه GRU پردازش می‌شوند. این لایه به صورت  bidirectional عمل می‌کند، به این معنی که ورودی‌ها از هر دو جهت (چپ به راست و راست به چپ) پردازش می‌شوند.
Fully Connected Layer: در نهایت، خروجی‌ها از طریق یک لایه خطی (fully connected) به ابعاد مناسب برای پیش‌بینی کلمات تبدیل می‌شوند.
مکانیزم توجه (Attention): اگر مکانیزم توجه فعال باشد، از کلاس Attention برای به دست آوردن وزن‌های توجه و ترکیب آنها با ورودی‌ها استفاده می‌شود.
سپس تابع optimizer تعریف شده است که از بهینه ساز adam برای بروزرسانی وزن های مدل استفاده می کند.
در مرحله بعد تابع loss را مشخص میکنیم که در اینجا از CrossEntropyLoss استفاده میشود.
سپس به سراغ آموزش و ارزیابی مدل می رویم:
مدل ابتدا برای هر ایپاک در حالت آموزش فرار میگیرد. سپس برای هر بچ از داده ها ورودی و خروجی ها به مدل داده شده و بروزرسانی وزدن ها با استفاده از گرادیان ها انجام می گیرد.
پس از محاسبه loss گرادیان ها و وزن های مدل بروز می شوند.
در مرحله آخر میانگین loss برای هر ایپاک محاسبه شده و ذخیره می شود.
در مرحله بعد پس از پایان هر ایپاک مدل در حالت ارزیابی قرار می گیرد. سپس مدل بر روی داده های  اعتبار سنجی تست می شود و میانگین loss برای داده های اعتبار سنجیی  محاسبه می شود و نتایج loss آموزش و اعتبار سنجی برای هر ایپاک چاپ میشود.

```python
import torch
import torch.nn as nn
import torch.optim as optim


class Attention(nn.Module):
    def __init__(self, hidden_size):
        super(Attention, self).__init__()
        self.attn = nn.Linear(hidden_size * 2, 1)  
        self.softmax = nn.Softmax(dim=1)

    def forward(self, encoder_outputs):
        attn_weights = self.softmax(self.attn(encoder_outputs))
        context = torch.sum(attn_weights * encoder_outputs, dim=1)
        return context, attn_weights

class Seq2Seq(nn.Module):
    def __init__(self, vocab, embedding_matrix, hidden_size):
        super(Seq2Seq, self).__init__()
     
        self.embedding = nn.Embedding.from_pretrained(torch.tensor(embedding_matrix, dtype=torch.float32), freeze=True)
      
        self.gru = nn.GRU(embedding_matrix.shape[1], hidden_size, batch_first=True, bidirectional=True)
      
        self.fc_out = nn.Linear(hidden_size * 2, vocab.vocab_size())  
    
    def forward(self, src, tgt):
     
        embedded = self.embedding(src)  
       
        output, hidden = self.gru(embedded)  
        
        output = self.fc_out(output) 
        
        return output
    



embedding_size = 100  
hidden_size = 100
batch_size = 32
num_epochs = 10
learning_rate = 1e-3


device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = Seq2Seq(vocab=vocab, embedding_matrix=embedding_matrix, hidden_size=hidden_size).to(device)

optimizer = optim.Adam(model.parameters(), lr=learning_rate)
criterion = nn.CrossEntropyLoss()

train_data_loader = DataLoader(Seq2SeqDataset(train_input_tokens, train_target_tokens, vocab), batch_size=batch_size, collate_fn=collate_fn)

val_data_loader = DataLoader(Seq2SeqDataset(val_input, val_target, vocab),
                             batch_size=32,  
                             collate_fn=collate_fn)

train_loss_list=[]
val_loss_list=[]
for epoch in range(num_epochs):
    model.train()      
    total_loss = 0
    
    for input_seqs, target_seqs in train_data_loader:
        input_seqs = input_seqs.to(device)
        target_seqs = target_seqs.to(device)
        
        optimizer.zero_grad() 
        
        outputs = model(input_seqs, target_seqs)
        
        outputs = outputs.view(-1, outputs.size(-1)) 
        target_seqs = target_seqs.view(-1)  
        
        loss = criterion(outputs, target_seqs)
        total_loss += loss.item()
        
        loss.backward()  
        optimizer.step()  
    
    average_loss = total_loss / len(train_data_loader)
    print(average_loss)
    train_loss_list.append(average_loss)
  
    model.eval()  
    val_loss = 0
    with torch.no_grad(): 
        for input_seqs, target_seqs in val_data_loader:
            input_seqs = input_seqs.to(device)
            target_seqs = target_seqs.to(device)
            
            outputs = model(input_seqs, target_seqs)
            
            outputs = outputs.view(-1, outputs.size(-1))  
            target_seqs = target_seqs.view(-1) 
            
            loss = criterion(outputs, target_seqs)
            val_loss += loss.item()
    
    average_val_loss = val_loss / len(val_data_loader)
    val_loss_list.append(average_val_loss)

    print(f"Epoch [{epoch + 1}/{num_epochs}], Train Loss: {average_loss:.4f}, Validation Loss: {average_val_loss:.4f}")
```

<h4 dir="rtl" style="font-family: Vazir; width: 85%;">صداقت علمی:</h4> <ul dir="rtl" style="font-family: Vazir; width: 85%; font-size: 16px;"> <li>ما نوت‌بوک‌های تعداد مشخصی از دانشجویان که به صورت تصادفی انتخاب می‌شوند، بررسی خواهیم کرد. این بررسی‌ها اطمینان حاصل می‌کنند که کدی که نوشتید واقعاً پاسخ‌های موجود در نوت‌بوک شما را تولید می‌کند. اگر پاسخ‌های صحیح را در نوت‌بوک خود بدون کدی که واقعاً آن پاسخ‌ها را تولید کند تحویل دهید، این یک مورد جدی از عدم صداقت علمی محسوب می‌شود.</li> <li>ما همچنین بررسی‌های خودکاری را برای تشخیص سرقت علمی در نوت‌بوک‌های کولب انجام خواهیم داد. کپی کردن کد از دیگران نیز یک مورد جدی از عدم صداقت علمی محسوب می‌شود.</li> </ul>

<h4 dir="rtl" style="font-family: Vazir; width: 85%;">توضیحات تکمیلی:</h4> <ul dir="rtl" style="font-family: Vazir; width: 85%; font-size: 16px;">
<li>
خوانایی و دقت بررسی‌ها در گزارش نهایی از اهمیت ویژه‌ای برخوردار است. به تمرین‌هایی که به صورت کاغذی تحویل داده شوند یا به صورت عکس در سایت بارگذاری شوند، ترتیب اثری داده نخواهد شد.</li>
<li>
 همه‌ی کدهای پیوست گزارش بایستی قابلیت اجرای مجدد داشته باشند. در صورتی که برای اجرا مجدد آن‌ها نیاز به تنظیمات خاصی می‌باشد، بایستی تنظیمات مورد نیاز را نیز در گزارش خود ذکر کنید.  دقت کنید که  تمامی کدها باید توسط شما اجرا شده باشند و نتایج اجرا در فایل کدهای ارسالی مشخص باشد. به کدهایی که نتایج اجرای آن‌ها در فایل ارسالی مشخص نباشد نمره‌ای تعلق نمی‌گیرد.
</li>
<li>
تمرین تا یک هفته بعد از مهلت تعیین شده با تاخیر تحویل گرفته می‌شود. دقت کنید که شما جمعاً برای تمام تکالیف، ۱۴ روز زمان تحویل بدون جریمه دارید که تنها از ۷ روز آن برای هر تمرین می‌توانید استفاده کنید. در صورتی که این ۱۴ روز به اتمام رسیده باشد، به ازای هر روز تاخیر ده درصد جریمه می‌شود.
