# پاسخ خود را اینجا بنویسید:

<div style="direction: rtl; text-align: right;">
    الف:
    <br><br>
    الگوریتم‌های Levenshtein Distance و Damerau-Levenshtein Distance ابزارهایی برای اندازه‌گیری شباهت بین دو رشته (string) هستند. این الگوریتم‌ها به‌ویژه در پردازش زبان طبیعی، اصلاح خودکار متن و شناسایی الگو کاربرد دارند. در ادامه، هر یک از این الگوریتم‌ها و طرز کار آن‌ها را شرح می‌دهم.
    <br><br>
    1. Levenshtein Distance
    <br><br>
    تعریف: Levenshtein Distance فاصله ویرایش (edit distance) بین دو رشته را اندازه‌گیری می‌کند. این فاصله نشان می‌دهد که حداقل چند عملیات ویرایش (حذف، اضافه، یا جایگزینی کاراکتر) برای تبدیل یک رشته به رشته دیگر لازم است.
    <br><br>
    طرز کار:
    <br><br>
    •	عملیات‌های مجاز عبارتند از:
    <br><br>
    1.	حذف: حذف یک کاراکتر از رشته.
    <br><br>
    2.	اضافه: اضافه کردن یک کاراکتر به رشته.
    <br><br>
    3.	جایگزینی: جایگزینی یک کاراکتر با کاراکتر دیگر.
    <br><br>
    2. Damerau-Levenshtein Distance
    <br><br>
    تعریف: Damerau-Levenshtein Distance نسخه‌ای بهبود یافته از Levenshtein Distance است که علاوه بر عملیات‌های سه‌گانه فوق، عمل جابجایی (transposition) دو کاراکتر مجاور را نیز در نظر می‌گیرد.
    <br><br>
    طرز کار:
    <br><br>
    عملیات‌های مجاز عبارتند از:
    <br><br>
    1.	حذف: حذف یک کاراکتر از رشته.
    <br><br>
    2.	اضافه: اضافه کردن یک کاراکتر به رشته.
    <br><br>
    3.	جایگزینی: جایگزینی یک کاراکتر با کاراکتر دیگر.
    <br><br>
    4.	جابجایی: جابجایی دو کاراکتر مجاور.
    <br><br>
    مقایسه:
    <br><br>
    Levenshtein Distance تنها سه عملیات ویرایش را در نظر می‌گیرد، در حالی که Damerau-Levenshtein Distance چهار عملیات ویرایش را در نظر می‌گیرد.
    <br><br>
    بنابراین، Damerau-Levenshtein Distance معمولاً دقت بیشتری در شناسایی شباهت‌ها و تفاوت‌ها بین رشته‌ها ارائه می‌دهد، به‌ویژه زمانی که جابجایی کاراکترها در متن وجود دارد.
    <br><br>
    کاربردها:
    <br><br>
    تصحیح خودکار متن: شناسایی و تصحیح خطاهای املایی.
    <br><br>
    جستجوی تقریب: جستجوی رشته‌هایی که به طور تقریبی مشابه یکدیگر هستند.
    <br><br>
    تحلیل متن: ارزیابی شباهت بین متن‌ها و شناسایی تغییرات.

</div>


```python
#write your code here

def levenshtein(str1, str2):
    len_str1 = len(str1) + 1 
    len_str2 = len(str2) + 1
    matrix = [[0 for _ in range (len_str2)] for _ in range (len_str1)]
    for i in range (len_str1):
        matrix [i][0] = i
    for j in range (len_str2):
        matrix [0][j] = j
    
    for i in range(1, len_str1):
        for j in range(1, len_str2):
            if str1[i-1] == str2[j-1]:
                cost = 0
            else:
                cost = 1
            matrix [i][j] = min(matrix[i-1][j] + 1,
                                matrix[i][j-1] + 1,
                                matrix[i-1][j-1] + cost)
    return matrix[len_str1-1][len_str2-1]




def damerau_levenshtein(str1, str2):
    len_str1 = len(str1) + 1 
    len_str2 = len(str2) + 1
    matrix = [[0 for _ in range (len_str2)] for _ in range (len_str1)]
    for i in range (len_str1):
        matrix [i][0] = i
    for j in range (len_str2):
        matrix [0][j] = j
    
    for i in range(1, len_str1):
        for j in range(1, len_str2):
            if str1[i-1] == str2[j-1]:
                cost = 0
            else:
                cost = 1
            matrix [i][j] = min(matrix[i-1][j] + 1,
                                matrix[i][j-1] + 1,
                                matrix[i-1][j-1] + cost)
            if i > 1 and j > 1 and str1[i - 1] == str2[j - 2] and str1[i - 2] == str2[j - 1]:
                matrix[i][j] = min(matrix[i][j], matrix[i - 2][j - 2] + 1) 
    return matrix[len_str1-1][len_str2-1]


list = [ ("kitten", "sitting"), ("saturday", "sunday"), ("book", "back"), ("algorithm",
"logarithm"), ("", "test"), ("abc", "acb") ]

for i in list:
    print('levenshtein distance between' ,i[0], 'and', i[1], 'is:', levenshtein(i[0], i[1]))


for i in list:
    print('damerau-levenshtein distance between' ,i[0], 'and', i[1], 'is:', damerau_levenshtein(i[0], i[1]))
```

<div style="direction: rtl; text-align: right;">
    Levenshtein
    <br><br>
    این تابع برای محاسبه فاصله Levenshtein بین دو رشته طراحی شده است. فاصله Levenshtein تعداد تغییرات (افزودن، حذف، یا جایگزینی یک کاراکتر) مورد نیاز برای تبدیل یک رشته به رشته دیگر را محاسبه می‌کند.
    <br><br>
    ابتدا طول دو رشته ورودی را به همراه ۱ به دست می‌آورد و یک ماتریس دو بعدی با ابعاد [طول رشته ۱ + ۱] در [طول رشته ۲ + ۱] ایجاد می‌کند. هر عنصر این ماتریس برای نگه‌داشتن کمترین هزینه تبدیل بین دو رشته استفاده می‌شود.
    <br><br>
    سطر و ستون اول ماتریس با اعداد از ۰ تا طول رشته‌ها پر می‌شوند، که نمایانگر هزینه تبدیل از یک رشته خالی به رشته مورد نظر است.
    <br><br>
    سپس برای هر کاراکتر دو رشته، بررسی می‌کند که آیا کاراکترهای مورد نظر برابر هستند یا خیر:
    <br><br>
    اگر برابر باشند، هزینه تبدیل ۰ است.
    <br>
    اگر برابر نباشند، هزینه ۱ به تبدیل اضافه می‌شود.
    <br><br>
    کمترین هزینه از سه عملیات (حذف، درج، یا جایگزینی) انتخاب شده و در ماتریس ذخیره می‌شود.
    <br><br>
    در پایان، آخرین مقدار ماتریس (پایین‌ترین راست) که نشان‌دهنده کمترین هزینه تبدیل کل دو رشته است، برگردانده می‌شود.
    <br><br><br>
    Damerau-Levenshtein
    <br><br>
    این تابع نیز فاصله بین دو رشته را محاسبه می‌کند، اما با یک تغییر مهم: در این متد، علاوه بر عملیات‌های درج، حذف، و جایگزینی، یک عملیات دیگر نیز مجاز است که به آن transposition می‌گویند. در این حالت، دو کاراکتر مجاور می‌توانند جابه‌جا شوند.
    <br><br>
    همانند تابع Levenshtein، اینجا نیز ابتدا ماتریس ایجاد می‌شود و سطر و ستون اول آن با اعداد پر می‌شود.
    <br><br>
    عملیات‌های مقایسه و هزینه مشابه هستند، اما اینجا یک بررسی اضافه وجود دارد: اگر کاراکترهای متوالی در دو رشته معادل هم باشند، اما ترتیب آن‌ها متفاوت باشد، می‌تواند یک هزینه کمتر (۱ واحد) برای جابه‌جایی آن‌ها در نظر بگیرد.
    <br><br>
    این ویژگی باعث می‌شود که این الگوریتم دقیق‌تر باشد، چرا که جابه‌جایی‌ها را به عنوان یک عملیات واحد در نظر می‌گیرد.

    الف. نحوه آموزش WordPiece Tokenizer
    <br><br>

    می‌شود. نحوه آموزش این توکنایزر به‌طور خلاصه به شرح زیر است:

# پاسخ خود را اینجا بنویسید:


<div style="direction: rtl; text-align: right;">
    ج
    <br><br>
    1. فضای ذخیره‌سازی بزرگ
    <br><br>
    با افزایش مقدار n n، تعداد ترکیب‌های ممکن از توکن‌ها به‌طرز چشمگیری افزایش می‌یابد. به‌عنوان مثال، در یکمجموعه داده بزرگ، ممکن است نیاز به ذخیره‌سازی میلیون‌ها یا حتی میلیاردها n-gram وجود داشته باشد. این امر به‌ویژه در مدل‌های بزرگ و پیچیده، به منابع حافظه زیادی نیاز دارد.
    <br><br>
    2. کمبود داده (Sparsity)
    <br><br>
    در n-gram های بزرگ، بسیاری از ترکیب‌ها ممکن است در داده‌های آموزشی وجود نداشته باشند. این کمبود داده می‌تواند منجر به مشکل "کمبود داده" (data sparsity) شود، که به این معناست که n-gram های خاصی ممکن است به‌اندازه کافی اطلاعات نداشته باشند تا بتوانند پیش‌بینی‌های دقیقی انجام دهند.
    <br><br>
    3. مدل‌سازی وابستگی‌های بلندمدت
    <br><br>
    n-gram ها معمولاً به روابط محلی بین توکن‌ها توجه دارند و در مدل‌سازی وابستگی‌های بلندمدت (long-range dependencies) ناتوان هستند. به همین دلیل، آن‌ها نمی‌توانند اطلاعات مربوط به توالی‌های طولانی‌تر را به‌خوبی پردازش کنند، که این مسأله می‌تواند در مواردی که متن به شدت وابسته است، منجر به نتایج نادرست شود.
    <br><br>
    4. عدم قطعیت و تنوع
    <br><br>
    n-gram ها قادر به مدل‌سازی عدم قطعیت در پیش‌بینی‌ها نیستند. به‌عنوان مثال، ممکن است یک ترکیب n-gram در یک متن مشخص وجود نداشته باشد، اما مدل بر اساس توکن‌های نزدیک‌تر، پیش‌بینی‌های نادرستی انجام دهد. این به این معناست که ن-گرام ها نمی‌توانند تنوع و تطبیق‌های مختلف در زبان را به‌خوبی در نظر بگیرند.
    <br><br>
    5. توزیع‌های غیر یکنواخت
    <br><br>
    در بسیاری از زبان‌ها، برخی از توکن‌ها به‌طور طبیعی بیشتر از دیگر توکن‌ها ظاهر می‌شوند. این توزیع‌های غیر یکنواخت می‌تواند منجر به بایاس (bias) در مدل شود و باعث شود که برخی از توکن‌ها یا n-gram ها به‌طور غیرمنصفانه‌ای بر نتایج تأثیر بگذارند.
    <br><br>
    6. هزینه محاسباتی
    <br><br>
    افزایش مقدار 𝑛n منجر به افزایش هزینه‌های محاسباتی می‌شود. در هنگام پیش‌بینی، مدل باید تمام n-gram های ممکن را بررسی کند که این فرآیند می‌تواند زمان‌بر باشد و منابع محاسباتی زیادی را مصرف کند.
    <br><br>
    7. محدودیت در پیش‌بینی
    <br><br>
    n-gram ها محدود به پیش‌بینی توکن‌ها بر اساس تعداد ثابتی از توکن‌های قبلی هستند. به همین دلیل، این روش‌ها در مسائلی مانند تولید متن یا تولید زبان طبیعی با تنوع و پیچیدگی بالا با چالش‌هایی روبه‌رو می‌شوند.
</div>


```python
#write your code here
#write your code here
import collections
import random
from tokenizers import Tokenizer

class NgramModel:
    def __init__(self, n, tokenizer_path):
        self.n = n
        self.ngrams = collections.defaultdict(collections.Counter)
        self.vocab = set()
        self.tokenizer = Tokenizer.from_file(tokenizer_path)

    def tokenize(self, text):
        return self.tokenizer.encode(text).tokens

    def build_ngrams(self, tokens):
        for i in range(len(tokens) - self.n + 1):
            gram = tuple(tokens[i:i + self.n - 1])
            next_word = tokens[i + self.n - 1]
            self.ngrams[gram][next_word] += 1
            self.vocab.add(next_word)

    def train(self, text):
        tokens = self.tokenize(text)
        self.build_ngrams(tokens)
        print(f"Trained on {len(tokens)} tokens, built {sum(len(v) for v in self.ngrams.values())} n-grams.")

    def predict(self, context):
        context = tuple(context)
        possible_next_words = self.ngrams.get(context, {})

        if possible_next_words:
            vocab_size = len(self.vocab)
            total_count = sum(possible_next_words.values()) + vocab_size  # Add-1 smoothing
            # Calculate smoothed probabilities
            words, probs = zip(*[(word, (count + 1) / total_count) for word, count in possible_next_words.items()])
            
            # Selecting next word based on probabilities
            return random.choices(words, weights=probs)[0]
        else:
            # Fallback to random choice if no context found
            return random.choice(list(self.vocab))  # Ensure list is not redefined

    def generate_text(self, seed_text, length):
        generated = seed_text[:]
        context = seed_text[-(self.n - 1):]
        for _ in range(length - len(seed_text)):
            next_word = self.predict(context)
            generated.append(next_word)
            context = generated[-(self.n - 1):]
        return " ".join(generated)

tokenizer_path = "/Users/erfanshahabi/Desktop/uni/NLP/CA1/wordpiece_tokenizer.json"

models = {
    "2-gram": NgramModel(2, tokenizer_path),
    "4-gram": NgramModel(4, tokenizer_path),
    "8-gram": NgramModel(8, tokenizer_path)
}

with open("/Users/erfanshahabi/Desktop/uni/NLP/CA1/NLP_CA1/data/ferdowsi.txt", "r", encoding='utf-8') as file:
    text = file.read()

for name, model in models.items():
    print(f"Training {name}...")
    model.train(text)

seed_text = ["هنر", "همانا"]  # بذر برای شروع متن
length = 200  # طول متن تولیدی

for name, model in models.items():
    print(f"\nGenerating text with {name} model:")
    generated_text = model.generate_text(seed_text, length)
    print(generated_text)

```

```python
#write your code here
#write your code here
import collections
import random
import math
from tokenizers import Tokenizer

class NgramModel:
    def __init__(self, n, tokenizer_path):
        self.n = n
        self.ngrams = collections.defaultdict(collections.Counter)
        self.vocab = set()
        self.tokenizer = Tokenizer.from_file(tokenizer_path)

    def tokenize(self, text):
        return self.tokenizer.encode(text).tokens

    def build_ngrams(self, tokens):
        for i in range(len(tokens) - self.n + 1):
            gram = tuple(tokens[i:i + self.n - 1])
            next_word = tokens[i + self.n - 1]
            self.ngrams[gram][next_word] += 1
            self.vocab.add(next_word)

    def train(self, text):
        tokens = self.tokenize(text)
        self.build_ngrams(tokens)
        print(f"Trained on {len(tokens)} tokens, built {sum(len(v) for v in self.ngrams.values())} n-grams.")

    def predict(self, context):
        context = tuple(context)
        possible_next_words = self.ngrams.get(context, {})

        if possible_next_words:
            vocab_size = len(self.vocab)
            total_count = sum(possible_next_words.values()) + vocab_size  # Add-1 smoothing
            # Calculate smoothed probabilities
            words, probs = zip(*[(word, (count + 1) / total_count) for word, count in possible_next_words.items()])
            
            # Selecting next word based on probabilities
            return random.choices(words, weights=probs)[0]
        else:
            # Fallback to random choice if no context found
            return random.choice(list(self.vocab))  # Ensure list is not redefined

    def generate_text(self, seed_text, length):
        generated = seed_text[:]
        context = seed_text[-(self.n - 1):]
        for _ in range(length - len(seed_text)):
            next_word = self.predict(context)
            generated.append(next_word)
            context = generated[-(self.n - 1):]
        return " ".join(generated)
    def perplexity(self, text):
        tokens = self.tokenize(text)
        if len(tokens) < self.n:
            return float('inf')  # اگر تعداد توکن‌ها کمتر از n باشد، perplexity بی‌نهایت می‌شود
        log_prob_sum = 0
        N = len(tokens) - self.n + 1

        for i in range(N):
            context = tuple(tokens[i:i + self.n - 1])
            next_word = tokens[i + self.n - 1]
            if next_word in self.ngrams[context]:
                count = self.ngrams[context][next_word]
                prob = (count + 1) / (sum(self.ngrams[context].values()) + len(self.vocab))  # Add-1 smoothing
                log_prob_sum += math.log(prob)
            else:
                log_prob_sum += math.log(1 / (sum(self.ngrams[context].values()) + len(self.vocab)))  # اگر کلمه‌ای در n-gram وجود نداشته باشد

        return math.exp(-log_prob_sum / N)  # محاسبه perplexity



tokenizer_path = "/Users/erfanshahabi/Desktop/uni/NLP/CA1/wordpiece_tokenizer.json"
model = NgramModel(4, tokenizer_path) 

with open("/Users/erfanshahabi/Desktop/uni/NLP/CA1/NLP_CA1/data/ferdowsi.txt", "r", encoding='utf-8') as file:
    text = file.read()

model.train(text)

hafez_perplexity = []  
with open("/Users/erfanshahabi/Desktop/uni/NLP/CA1/NLP_CA1/data/hafez.txt", "r", encoding='utf-8') as file:
    for line in file:
        perplexity = model.perplexity(line.strip())
        hafez_perplexity.append(perplexity)

mean_hafez_perplexity = sum(hafez_perplexity) / len(hafez_perplexity)
print(f"Mean Perplexity for Hafez: {mean_hafez_perplexity}")

new_poems_perplexity = []
with open("/Users/erfanshahabi/Desktop/uni/NLP/CA1/NLP_CA1/data/modern_poet.txt", "r", encoding='utf-8') as file:
    for line in file:
        perplexity = model.perplexity(line.strip())
        new_poems_perplexity.append(perplexity)

mean_new_poems_perplexity = sum(new_poems_perplexity) / len(new_poems_perplexity)
print(f"Mean Perplexity for Modern Poems: {mean_new_poems_perplexity}")


print(f"Comparison: Hafez vs Modern Poems")
print(f"Hafez Perplexity: {mean_hafez_perplexity}")
print(f"Modern Poems Perplexity: {mean_new_poems_perplexity}")
```

```python
#write your code here
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, precision_score, recall_score
from tokenizers import Tokenizer
from sklearn.feature_extraction.text import TfidfVectorizer

class NgramClassifier:
    def __init__(self, n, tokenizer_path):
        self.tokenizer = Tokenizer.from_file(tokenizer_path)  
        self.vectorizer = TfidfVectorizer(ngram_range=(n, n), max_df=0.95, min_df=2, stop_words='english')
        self.model = MultinomialNB()

    def tokenize_text(self, text):
        return " ".join(self.tokenizer.encode(text).tokens) 

    def train(self, X_train, y_train):
        X_train_tokenized = X_train.apply(self.tokenize_text)
        X_train_counts = self.vectorizer.fit_transform(X_train_tokenized)
        self.model.fit(X_train_counts, y_train)

    def evaluate(self, X_test, y_test):
        X_test_tokenized = X_test.apply(self.tokenize_text)
        X_test_counts = self.vectorizer.transform(X_test_tokenized)
        y_pred = self.model.predict(X_test_counts)
        acc = accuracy_score(y_test, y_pred)
        precision = precision_score(y_test, y_pred, average='weighted', zero_division=1)
        recall = recall_score(y_test, y_pred, average='weighted')
        return acc, precision, recall

data = pd.read_csv('/Users/erfanshahabi/Desktop/uni/NLP/CA1/NLP_CA1/data/digikala.csv')

data.dropna(subset=['Text', 'Suggestion'], inplace=True)
train_data, test_data = train_test_split(data, test_size=0.2, random_state=42)
X_train, y_train = train_data['Text'], train_data['Suggestion']
X_test, y_test = test_data['Text'], test_data['Suggestion']

ngram_2 = NgramClassifier(2, "/Users/erfanshahabi/Desktop/uni/NLP/CA1/bpe_tokenizer.json")
ngram_2.train(X_train, y_train)
acc_2, precision_2, recall_2 = ngram_2.evaluate(X_test, y_test)

ngram_3 = NgramClassifier(3, "/Users/erfanshahabi/Desktop/uni/NLP/CA1/bpe_tokenizer.json")
ngram_3.train(X_train, y_train)
acc_3, precision_3, recall_3 = ngram_3.evaluate(X_test, y_test)


print(f"2-gram - Accuracy: {acc_2}, Precision: {precision_2}, Recall: {recall_2}")
print(f"3-gram - Accuracy: {acc_3}, Precision: {precision_3}, Recall: {recall_3}")

if acc_2 > acc_3:
    print("مدل 2-gram دقت بهتری دارد.")
else:
    print("مدل 3-gram دقت بهتری دارد.")
```
