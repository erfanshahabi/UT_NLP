در این قسمت از کد پیش پردازش داده ها، پد کردن و ادغام آن ها صورت گرفته است.
ابتدا تابع create_ner_tag_mapping تعریف شده است که به ورودی داده ها نگاه میکند و یک نگاشت از لیبل های عددی به نام لیبل ها ایچاد می کند. این نگاشت ها به صورت دو دیکشنری ذخیره میشوند:
id2label: نگاشت از برچسب عددی به نام برچسب متنی.
label2id: نگاشت از نام برچسب متنی به برچسب عددی.
سپس این نگاشت‌ها برای داده‌های فارسی و انگلیسی استخراج می‌شود.
سپس تابع پیش پردازش توکن ها نوشته شده است که در این تابع توکن های ورودی به حروف کوچک تبدیل می شوند و نشانه گذاری ها و علائم اضافی نیز حذف می گردند.
در مرحله بعد به سراغ پیش پردازش داده ها می رویم.
در این قسمت برای هر داده لیبل های ner_tags به صورت عددی و لیبل های متنی ner_tags_names ایجاد می شوند و در صورت وجود ویژگی های دیگری مانند pos_tags , chunk_tags نیز اضافه مشوند
سپس این تابع برای داده های فارسی و انگلیسی اعمال می شود و داده های مربوط به هر زبان پردازش می شوند.
پس از پیش پردازش، داده های پردازش شده فارسی و انگلیسی برای هر بخش آموزش و تست و اعتبار سنجی ترکیب میشوند. سپس این داده ها به فرمت pandas.DataFrame تبدیل شده و به صورت مجموعه داده dataset از کتابخانه datasets تبدیل میشوند.
در مرحله بعد به سراغ پر کردن توکن ها میرویم
در این مرحله توکن ها و برچسب های مربوط به آن ها پد می شوند تا طول ثابت و یکسانی پیدا کنند.
برای این کار به این صورت عمل میکنیم که اگر طول توکن ها کنتر از مفدار ماکزیمم باشد توکن های اضافی با مقدار pad اضافه می شوند و برای برچسب ها اگر تعدادشان کمتر از طول توکن ها باشد برچسب ۰ اضافه می شود و این عملیات برای تمام داده ها به صورت بچ انجام می شود.
سپس به سراغ ترکیب داده های نهایی میرویم و داده ها در بک دیکشنری DatasetDivt ذخیره می شوند که شامل سه بخش داده های ترین تست و اعتبار سنجی است. در انتها داد های نهایی پردازش شده نمایش داده می شوند و  دو نگاشت id2label و label2id برای برچسب‌های NER چاپ می‌شون.

```python
import pandas as pd
from datasets import Dataset, DatasetDict
from transformers import AutoTokenizer

def create_ner_tag_mapping(dataset):
    id2label = {}
    label2id = {}
    for example in dataset:
        if 'ner_tags' in example and 'ner_tags_names' in example:
            for tag, tag_name in zip(example['ner_tags'], example['ner_tags_names']):
                if tag not in id2label:
                    id2label[tag] = tag_name
                    label2id[tag_name] = tag
    return id2label, label2id


persian_id2label, persian_label2id = create_ner_tag_mapping(persian_dataset['train'])
english_id2label, english_label2id = create_ner_tag_mapping(english_dataset['train'])


all_id2label = {**persian_id2label, **english_id2label}
all_label2id = {**persian_label2id, **english_label2id}


print("id2label:", all_id2label)
print("label2id:", all_label2id)


def preprocess_text(tokens):

    cleaned_tokens = [token.lower() for token in tokens if token.isalnum()]
    return cleaned_tokens


def preprocess_dataset(dataset, language, additional_features=None):
    processed_data = []
    for example in dataset:
        tokens = example['tokens']
        processed_tokens = preprocess_text(tokens)


        data_dict = {'tokens': processed_tokens, 'language': language}


        if 'ner_tags' in example:

            ner_tags = example['ner_tags']
            ner_tags_names = [all_id2label.get(tag, "O") for tag in ner_tags]
            data_dict['ner_tags'] = ner_tags
            data_dict['ner_tags_names'] = ner_tags_names
        if 'pos_tags' in example:
            data_dict['pos_tags'] = example['pos_tags']
        if 'chunk_tags' in example:
            data_dict['chunk_tags'] = example['chunk_tags']

        processed_data.append(data_dict)
    return processed_data


persian_train_data = preprocess_dataset(persian_dataset['train'], 'persian', additional_features=['ner_tags_names'])
persian_valid_data = preprocess_dataset(persian_dataset['validation'], 'persian', additional_features=['ner_tags_names'])
persian_test_data = preprocess_dataset(persian_dataset['test'], 'persian', additional_features=['ner_tags_names'])

english_train_data = preprocess_dataset(english_dataset['train'], 'english', additional_features=['pos_tags', 'chunk_tags', 'ner_tags'])
english_valid_data = preprocess_dataset(english_dataset['validation'], 'english', additional_features=['pos_tags', 'chunk_tags', 'ner_tags'])
english_test_data = preprocess_dataset(english_dataset['test'], 'english', additional_features=['pos_tags', 'chunk_tags', 'ner_tags'])


train_data = persian_train_data + english_train_data
valid_data = persian_valid_data + english_valid_data
test_data = persian_test_data + english_test_data


train_dataset = Dataset.from_pandas(pd.DataFrame(train_data))
valid_dataset = Dataset.from_pandas(pd.DataFrame(valid_data))
test_dataset = Dataset.from_pandas(pd.DataFrame(test_data))


dataset = DatasetDict({
    'train': train_dataset,
    'validation': valid_dataset,
    'test': test_dataset
})


def pad_tokens_and_labels(example, max_length=None):
    tokens = example['tokens']
    labels = example['ner_tags']

    if max_length is None:
        max_length = max(len(tokens), len(labels))


    padded_tokens = tokens + ['[PAD]'] * (max_length - len(tokens))

    padded_labels = labels + ['O'] * (max_length - len(labels))

    example['tokens'] = padded_tokens
    example['ner_tags'] = padded_labels

    return example


dataset = dataset.map(lambda examples: pad_tokens_and_labels(examples), batched=True)

print(dataset)
```

در این کد سعی شده از مدل XLMRoberta استفاده شود تا بتوانیم موجودیت های نام دار در یک متن را شناسایی کنیم.
ابتدا کلاس XLMRTokenClassification تعریف شده است که این کلاس یک مدل سفارشی برای لیبل گذاری توکنی است. در این کد از مدل از پیش آموزش شده XLMRoberta به عنوان بک بون استفاده میشود.
در ادامه یک لایه خطی به عنوان کلاسیفایر برای پیش بینی لیبل ها اضافه شده و از چک پوینت گرادیان برای کاهش حافظه استفاده می شود.
در ادامه مدل XLMRobertaForTokenClassification با ۲۱ لیبل بارگذاری شده و چک پوینت گیری گرادیان برای مدل فعال شده است. همچینین از XLMRobertaTokenizerFast برای توکنایز کردن داده ها استفاده می شود.
در مرحله بعد داده ها با استفاده از تابع tokenize_and_labels له توکن های ورودی تبدیل می شوند و لیبل ها به توکن ها نگاشت داده شده و توکن های خاص با مقدار -۱۰۰ نشان داده می شوند تا در محاسبه خطا نادیده گرفته شوند.   
سپس پارامترهای آموزش تعریف می شوند و داده ها به مجموعه های ترین و ولیدیشن تقسیم می شوند. در ادامه ترینر ایجاد شده و مدل شروع به آموزش می کند و پس از آموزش مدل روی داده های ولیدیشن ست شده و نتایج با trainer.evaluate() محاسبه میشوند. نتایج ارزیابی مدل در متغیر results ذخیره شده و شامل معیارهایی مانند دقت، خطا، و بازده می‌باشد.

```python
import torch
from torch import nn
from transformers import XLMRobertaModel, XLMRobertaTokenizer, TrainingArguments, Trainer, DataCollatorForTokenClassification
from transformers import AutoTokenizer
from transformers import XLMRobertaModel
from torch.utils.checkpoint import checkpoint

from transformers import BertTokenizerFast


num_labels = 21

class XLMRTokenClassification(nn.Module):
    def __init__(self, model_name, num_labels):
        super().__init__()
        self.backbone = XLMRobertaModel.from_pretrained(model_name)
        self.classifier = nn.Linear(self.backbone.config.hidden_size, num_labels)

    def forward(self, input_ids, attention_mask=None, labels=None):

        def custom_forward(*inputs):
            outputs = self.backbone(*inputs)
            return outputs.last_hidden_state

        sequence_output = checkpoint(custom_forward, input_ids, attention_mask)

        logits = self.classifier(sequence_output)

        loss = None
        if labels is not None:
            loss_fn = nn.CrossEntropyLoss()
            active_loss = attention_mask.view(-1) == 1
            active_logits = logits.view(-1, logits.shape[-1])[active_loss]
            active_labels = labels.view(-1)[active_loss]
            loss = loss_fn(active_logits, active_labels)

        return loss, logits

from transformers import XLMRobertaForTokenClassification, XLMRobertaTokenizer


model = XLMRobertaForTokenClassification.from_pretrained("xlm-roberta-base", num_labels=num_labels)


model.gradient_checkpointing_enable()

from transformers import XLMRobertaTokenizerFast


tokenizer = XLMRobertaTokenizerFast.from_pretrained("xlm-roberta-base")




def tokenize_and_labels(examples, tokenizer):

    tokenized_inputs = tokenizer(examples["tokens"], truncation=True, is_split_into_words=True, padding=True)


    labels = []
    for i, label in enumerate(examples["ner_tags"]):
        word_ids = tokenized_inputs.word_ids(batch_index=i)
        previous_word_idx = None
        label_ids = []
        for word_idx in word_ids:
            if word_idx is None:
                label_ids.append(-100)
            elif word_idx != previous_word_idx:
                label_ids.append(label[word_idx])
            else:
                label_ids.append(-100)
            previous_word_idx = word_idx
        labels.append(label_ids)


    tokenized_inputs["labels"] = labels
    return tokenized_inputs
num_labels = 21


tokenized_dataset = dataset.map(lambda x: tokenize_and_labels(x, tokenizer), batched=True)

train_dataset = tokenized_dataset["train"]
valid_dataset = tokenized_dataset["validation"]
import os
os.environ["WANDB_DISABLED"] = "true"

training_args = TrainingArguments(
    output_dir="Roberta-fa-en-ner", log_level="error",
    num_train_epochs=1,
    gradient_checkpointing=True,
    eval_accumulation_steps=10,
    per_device_train_batch_size=24,
    per_device_eval_batch_size=24,
    seed=42,
    logging_strategy="steps", eval_strategy="steps",
    save_steps=1e6, weight_decay=0.01, disable_tqdm=False,
    logging_steps=1e6, eval_steps=400, push_to_hub=False)


trainer = Trainer(
    model=model,
    args=training_args,
    data_collator=DataCollatorForTokenClassification(tokenizer),
    train_dataset=train_dataset,
    eval_dataset=valid_dataset,
    tokenizer=tokenizer
)


trainer.train()


results = trainer.evaluate()
print(results)
```

در ادامه این سوال این کد به منظور محاسبه معیار های ارزیابی precision, accuracy, recall , F1 score نوشته شده است.
ابتدا تابع compute_metrics رو تعریف کرده ام که این تابع وظیفه محاسبه معیار های ارزیابی را دارد. ورودی این تابع پیش بینی های مدل و لیبل های واقعی در نظر گرفته شده است و خروجی آن یک دیکشنری شامل معیار های ارزیابی است.
فرایند عملکرد تابع به این صورت است که از np.aargmax برای انتخاب برچسب با بالاترین احتمال استفاده می شود و مانند کد قبلی برچسب هایی که مقدارشان -۱۰۰ است یعنی آن هایی که لیبل های خاص هستند حذف می شوند تا در محاسبه معیار ها لحاظ نشوند. با استفاده از precision_recall_fscore_support معیارهای Precision، Recall، و F1-Score محاسبه می‌شوند و accuracy نیز با استفاده از accuracy_score محاسبه می شود. سپس trainer ایجاد شده و پارامترهای اصلی آن شامب مدل، داده ها، توکنایزر و تابع محاسبه معیار ها تنظیم می شوند.
در مرحله بعد فرایند اموزش مدل روی دادی های ترین انجام میشود و در دامه مدل روی داده های ولیدیشن تست می گردد و نتایح معیار ها در متغیر result ذخیره شده و نمایش داده می شود.
دلیل خطا برای پیش بینی تگ های جمله دلخواه بود که من اون قسمت رو پاک کردم و در ادامه برایش کدی جدا نوشتم ولی چون ران کردن کد زمان بر بود دوباره کد رو ران نکردم تا خطا پاک بشه.


```python

from sklearn.metrics import precision_recall_fscore_support, accuracy_score
import numpy as np

def compute_metrics(pred):
    predictions, labels = pred

    predictions = np.argmax(predictions, axis=2)

    true_labels = []
    true_predictions = []

    for prediction, label in zip(predictions, labels):
        for pred, lbl in zip(prediction, label):
            if lbl != -100:
                true_labels.append(lbl)
                true_predictions.append(pred)

    # محاسبه متریک‌ها
    precision, recall, f1, _ = precision_recall_fscore_support(true_labels, true_predictions, average="weighted")
    accuracy = accuracy_score(true_labels, true_predictions)

    return {
        "accuracy": accuracy,
        "precision": precision,
        "recall": recall,
        "f1": f1
    }

trainer = Trainer(
    model=model,
    args=training_args,
    data_collator=DataCollatorForTokenClassification(tokenizer),
    train_dataset=train_dataset,
    eval_dataset=valid_dataset,
    tokenizer=tokenizer,
    compute_metrics=compute_metrics
)


trainer.train()

results = trainer.evaluate()
print("Evaluation Results:")
print(results)


```

```python
import torch

def predict_example(sentence, tokenizer, model):

    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

    model.to(device)


    inputs = tokenizer(sentence, return_tensors="pt", truncation=True, padding=True)

    inputs = {key: value.to(device) for key, value in inputs.items()}


    model.eval()
    with torch.no_grad():
        outputs = model(**inputs)


    logits = outputs.logits
    predictions = torch.argmax(logits, dim=2)


    tokens = tokenizer.convert_ids_to_tokens(inputs["input_ids"].squeeze())
    predictions = predictions.squeeze().tolist()
    predicted_labels = [model.config.id2label[pred] for pred in predictions]


    for token, label in zip(tokens, predicted_labels):
        print(f"{token:<15} -> {label}")


example_sentence = "امیر حسین اهل ایران بود اما خانواده او متولد دانمارک بودند"


predict_example(example_sentence, tokenizer, model)
```

```python

from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig
from datasets import load_dataset
from rouge_score import rouge_scorer
from bert_score import score as bert_score
import torch
import evaluate



quantization_config = BitsAndBytesConfig(load_in_4bit=True)
checkpoint = "HuggingFaceTB/SmolLM-1.7B"

device = "cuda" if torch.cuda.is_available() else "cpu"
tokenizer = AutoTokenizer.from_pretrained(checkpoint)

tokenizer.pad_token = tokenizer.eos_token


model = AutoModelForCausalLM.from_pretrained(checkpoint, quantization_config=quantization_config)


def summarize_article(article_text, max_length=51, num_beams=5):
    prompt = f"Summarize this article maximum in 70 tokens: {article_text}"

    inputs = tokenizer(prompt, return_tensors="pt", truncation=True, max_length=1924, padding=True)

    attention_mask = inputs.attention_mask
    pad_token_id = tokenizer.pad_token_id if tokenizer.pad_token_id is not None else tokenizer.eos_token_id


    outputs = model.generate(
        inputs.input_ids,
        attention_mask=attention_mask,
        max_new_tokens=max_length,
        num_beams=num_beams,
        temperature=0.7,
        no_repeat_ngram_size=2,
        eos_token_id=tokenizer.eos_token_id,
        pad_token_id=pad_token_id
    )


    summary = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return summary


num_samples = 1


rouge = evaluate.load('rouge')
bertscore = evaluate.load('bertscore')

generated_summaries = []
reference_summaries = []

for idx, sample in enumerate(test_data.select(range(num_samples))):
    article = sample['article']
    reference_summary = sample['highlights']
    generated_summary = summarize_article(article, max_length=100)

    print(f"\n### sample {idx+1} ###")
    print("true summary:")
    print(reference_summary)
    print("\ngenerated summary:")
    print(generated_summary)


    reference_summaries.append(reference_summary)
    generated_summaries.append(generated_summary)

rouge_scores = rouge.compute(predictions=generated_summaries, references=reference_summaries)
print("\n### ROUGE scores ###")
for key, value in rouge_scores.items():
    print(f"{key}: {value:.4f}")


bertscore_results = bertscore.compute(predictions=generated_summaries, references=reference_summaries, lang="en")
print("\n### BERTScore ###")
print(f"Precision: {sum(bertscore_results['precision']) / len(bertscore_results['precision']):.4f}")
print(f"Recall: {sum(bertscore_results['recall']) / len(bertscore_results['recall']):.4f}")
print(f"F1: {sum(bertscore_results['f1']) / len(bertscore_results['f1']):.4f}")
```

به طور کلی، نتایج ROUGE نشان می‌دهد که مدل قادر به تولید خلاصه‌های دقیق و معنی‌دار نیست.

معیار recall برای این مدل برابر 0.8058 است که این مقدار نشان می دهد مدل خدود ۸۱٪ از کلمات مبهم و مرتبط موجود در خلاصه اصلی را بازیابی کرده است. این مقدار نشان دهنده این موضوع است که مدل توانسته بخشی از اطلاعات مهم خلاصه ها را دریافت کند.
امتیاز f1 که میانگینی از precision و recall است در اینجا حدود ۷۸٪ بوده که نشان میدهد مدل توانسته در توازن بین precision , recall نسبتا عملکرد خوبی داشته باشد.
به طور کلی میتوان گفت که در ROUGE، مدل نتواسته است تطابق خوبی بین خلاصه‌های تولیدی و مرجع برقرار کند، به‌ویژه برای ROUGE-2 که نشان‌دهنده تطابق جملات دوکلمه‌ای است.
در BERTScore، نتایج بهتری مشاهده می‌شود که نشان‌دهنده انتخاب صحیح‌تر کلمات و ساختار جملات است، ولی هنوز به عملکرد مطلوب نرسیده‌ایم.
برای مقایسه بین این دو مدل میتوان گفت متد zero shot عملکرد به مراتب بهتری نسبت به مدل few shot داشته است هرچند که عملکرد مدل با سناریو zero shot هم چندان بالا نبوده و مدل نتوانسته است ارتباط معنایی بین کلمات را در خلاصه تولیدی رعابت کند.


```python
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig
from datasets import load_dataset
import torch
import random


quantization_config = BitsAndBytesConfig(load_in_4bit=True)
checkpoint = "HuggingFaceTB/SmolLM-1.7B"

device = "cuda" if torch.cuda.is_available() else "cpu"
tokenizer = AutoTokenizer.from_pretrained(checkpoint)

tokenizer.pad_token = tokenizer.eos_token

model = AutoModelForCausalLM.from_pretrained(checkpoint, quantization_config=quantization_config)


def summarize_with_fewshot(train_data, test_article, max_length=51, num_beams=5):

    few_shot_examples = random.sample(list(train_data), 3)
    prompt = "Here are some examples of articles and their summaries:\n\n"
    for example in few_shot_examples:
        prompt += f"Article: {example['article']}\nSummary: {example['highlights']}\n\n"

    prompt += f"Now, summarize this article:\n{test_article}\n"

    inputs = tokenizer(prompt, return_tensors="pt", truncation=True, max_length=1924, padding=True)

    outputs = model.generate(
        inputs.input_ids.to(device),
        attention_mask=inputs.attention_mask.to(device),
        max_new_tokens=max_length,
        num_beams=num_beams,
        temperature=0.7,
        no_repeat_ngram_size=2,
        eos_token_id=tokenizer.eos_token_id,
        pad_token_id=tokenizer.pad_token_id
    )

    generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    summary_start = generated_text.find("Now, summarize this article:") + len("Now, summarize this article:\n")
    summary = generated_text[summary_start:].strip()
    return summary

num_samples = 1

from rouge_score import rouge_scorer
from bert_score import score as bert_score

rouge_scorer = rouge_scorer.RougeScorer(["rouge1", "rouge2", "rougeL"], use_stemmer=True)

reference_summaries = []
generated_summaries = []

for idx, sample in enumerate(test_data.select(range(num_samples))):
    article = sample['article']
    reference_summary = sample['highlights']
    generated_summary = summarize_with_fewshot(train_data, article, max_length=100)

    print(f"\n### sample {idx+1} ###")
    print("article:")
    print(article)
    print("\ntrue summary:")
    print(reference_summary)
    print("\ngenerated summary:")
    print(generated_summary)

    reference_summaries.append(reference_summary)
    generated_summaries.append(generated_summary)


rouge_scores = {"rouge1": 0, "rouge2": 0, "rougeL": 0}
for ref, gen in zip(reference_summaries, generated_summaries):
    scores = rouge_scorer.score(ref, gen)
    for key in rouge_scores.keys():
        rouge_scores[key] += scores[key].fmeasure

num_eval_samples = len(reference_summaries)
for key in rouge_scores.keys():
    rouge_scores[key] /= num_eval_samples

print("\n### ROUGE scores ###")
for key, value in rouge_scores.items():
    print(f"{key}: {value:.4f}")


P, R, F1 = bert_score(generated_summaries, reference_summaries, lang="en")
print("\n### BERTScore ###")
print(f"Precision: {P.mean().item():.4f}")
print(f"Recall: {R.mean().item():.4f}")
print(f"F1: {F1.mean().item():.4f}")
```

ابتدا داده ها توکنایز می شوند و وقالات ورودی با حداکثر طول مشخص توکنایز شده و به input_ids تبدیل می شوند. truncation , padding این اطمینان را میدهند که تمام داده ها طول ثابت داشته باشند.

در مرحله بعد به سراغ تنظیمات آموزش رفته ام و پارامتر های آمورش با استفاده از SFTconfig تنظیم شده اند و از دیتا کولیتور برای دسته بندی داده ها استفاده کرده ام.

```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, Trainer, TrainingArguments
from datasets import load_dataset
import numpy as np
from torch.utils.data import DataLoader
from trl import SFTConfig, SFTTrainer
from google.colab import drive


drive.mount('/content/drive')

device = "cuda" if torch.cuda.is_available() else "cpu"


dataset_split = {
    'train': train_data,
    'validation': val_data,
    'test': test_data
}

max_input_length = 512
max_summary_length = 256


model_name = "HuggingFaceTB/SmolLM-360M"
tokenizer = AutoTokenizer.from_pretrained(model_name)

tokenizer.pad_token = tokenizer.eos_token


model = AutoModelForCausalLM.from_pretrained(model_name)
model.to(device)

def preprocess_function(examples):

    inputs = tokenizer(
        examples['article'],
        truncation=True,
        padding="max_length",
        max_length=max_input_length,
        return_tensors="pt"
    )


    summaries = tokenizer(
        examples['highlights'],
        truncation=True,
        padding="max_length",
        max_length=max_summary_length,
        return_tensors="pt"
    )


    inputs['labels'] = summaries['input_ids']

    inputs['labels'] = inputs['labels'].squeeze(0)
    print(f"Input shape: {inputs['input_ids'].shape}, Label shape: {inputs['labels'].shape}")

    return inputs


train_data = train_data.map(preprocess_function, batched=True)
val_data = val_data.map(preprocess_function, batched=True)
test_data = test_data.map(preprocess_function, batched=True)


output_dir = "/content/drive/MyDrive/output"
training_args = SFTConfig(
    output_dir=output_dir,
    num_train_epochs=1,
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    save_steps=1000,
    save_total_limit=200,
    logging_dir="./logs",
    fp16=True,
)

from transformers import DataCollatorForLanguageModeling


data_collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer,
    mlm=False
)


trainer = SFTTrainer(
    model=model,
    args=training_args,
    train_dataset=train_data,
    eval_dataset=val_data,
    data_collator=data_collator,
)

trainer.train()
```

 همه‌ی کدهای پیوست گزارش بایستی قابلیت اجرای مجدد داشته باشند. در صورتی که برای اجرا مجدد آن‌ها نیاز به تنظیمات خاصی می‌باشد، بایستی تنظیمات مورد نیاز را نیز در گزارش خود ذکر کنید.  دقت کنید که  تمامی کدها باید توسط شما اجرا شده باشند و نتایج اجرا در فایل کدهای ارسالی مشخص باشد. به کدهایی که نتایج اجرای آن‌ها در فایل ارسالی مشخص نباشد نمره‌ای تعلق نمی‌گیرد.
</li>
<li>
تمرین تا یک هفته بعد از مهلت تعیین شده با تاخیر تحویل گرفته می‌شود. دقت کنید که شما جمعاً برای تمام تکالیف، ۱۴ روز زمان تحویل بدون جریمه دارید که تنها از ۷ روز آن برای هر تمرین می‌توانید استفاده کنید. در صورتی که این ۱۴ روز به اتمام رسیده باشد، به ازای هر روز تاخیر ده درصد جریمه می‌شود.
</li>
